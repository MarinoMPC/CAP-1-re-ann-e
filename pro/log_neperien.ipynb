```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
Vmin,Vmax=0.1,4
nb_points=10000


dV=(Vmax-Vmin)/nb_points
V=np.linspace(Vmin,Vmax,nb_points)


def ln(V):
    return(np.log(V))

plt.plot(V,ln(V))
plt.grid()
```

```python
def derivee(fonction, variable, pas):
    f_V=fonction(variable)
    return((f_V[1:]-f_V[:-1])/pas)


f_prime=derivee(ln,V,dV)
```

```python
plt.plot(V[:-1],f_prime)
plt.grid()
```
